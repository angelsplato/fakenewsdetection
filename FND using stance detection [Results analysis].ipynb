```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#configurations
# load data set
directory = 'results/' 
```

```python
#High learning rate accuracy
higher_learning_rate_train_tag_epoch_accuracy = pd.read_csv(directory+'run-FND-higher-learning-rate_train-tag-epoch_accuracy.csv')
higher_learning_rate_validation_tag_epoch_accuracy= pd.read_csv(directory+'run-FND-higher-learning-rate_validation-tag-epoch_accuracy.csv')
plt.plot(higher_learning_rate_train_tag_epoch_accuracy['Step'] , higher_learning_rate_train_tag_epoch_accuracy['Value'], label='train')
plt.plot(higher_learning_rate_validation_tag_epoch_accuracy['Step'] , higher_learning_rate_validation_tag_epoch_accuracy['Value'], label='validation')
plt.title('High learning rate accuracy[L =0.5]')
plt.ylabel('accuracy value')
plt.xlabel('epoch')
plt.legend()
plt.savefig('highlearningrateaccuracy.png')
plt.show()
```

```python
#Low learning rate accuracy
lower_learning_rate_train_tag_epoch_accuracy = pd.read_csv(directory+'run-FND-lower-learning-rate_train-tag-epoch_accuracy.csv')
lower_learning_rate_validation_tag_epoch_accuracy= pd.read_csv(directory+'run-FND-lower-learning-rate_validation-tag-epoch_accuracy.csv')
plt.plot(lower_learning_rate_train_tag_epoch_accuracy['Step'] , lower_learning_rate_train_tag_epoch_accuracy['Value'], label='train')
plt.plot(lower_learning_rate_validation_tag_epoch_accuracy['Step'] , lower_learning_rate_validation_tag_epoch_accuracy['Value'], label='validation')
plt.title('Lower learning rate accuracy[L= 0.01]')
plt.ylabel('accuracy value')
plt.xlabel('epoch')
plt.legend()
plt.savefig('lowerlearningrateaccuracy.png')
plt.show()
```

```python
#Token[40 ] accuracy
token_40_train_tag_epoch_accuracy = pd.read_csv(directory+'run-FND-token-40_train-tag-epoch_accuracy.csv')
token_40_validation_tag_epoch_accuracy= pd.read_csv(directory+'run-FND-token-40_validation-tag-epoch_accuracy.csv')
plt.plot(token_40_train_tag_epoch_accuracy['Step'] , token_40_train_tag_epoch_accuracy['Value'], label='train')
plt.plot(token_40_validation_tag_epoch_accuracy['Step'] , token_40_validation_tag_epoch_accuracy['Value'], label='validation')
plt.title('Accuracy [Token length = 40]')
plt.ylabel('accuracy value')
plt.xlabel('epoch')
plt.legend()
plt.savefig('tokenlength=40.png')
plt.show()
```
